##### 6. Use your findings up to this point to create a multiple regression model with the following characteristics.

* Use at least 2 quantitative predictors, but do NOT use **invoice** as a predictor.
* Use one categorical predictor.

Remember to show the model summary.

```r
#Put solution to Exercise 6 in this cell. You may add additional cells if you like.
car_data$Sport <- as.integer(car_data$Body == "Sport")

car_sport <- car_data[car_data$Sport == 1,]
car_nonsport <- car_data[car_data$Sport == 0,]

model_eng <- lm(MSRP ~ EngineSize , data=car_sport)
model_eng1 <- lm(MSRP ~ EngineSize , data=car_nonsport)

shapes = c(16,17)
colors = c("red", "blue")
plot(MSRP ~ EngineSize, data = car_data, main="MSRP by Body ",
    col=colors[factor(car_data$Sport)], pch=shapes[factor(car_data$Sport)])

abline(model_eng,col="blue",lwd=2)
abline(model_eng1,col="red",lwd=2)

legend("topleft",
       legend = c("MSRP", "Engine"),
       pch = shapes,
       col = colors)

summary(model_eng)
summary(model_eng1)


```

```r
model_w <- lm(MSRP ~ Weight, data=car_sport)
model_h <- lm(MSRP ~ Weight , data=car_nonsport)


shapes = c(16,17)
colors = c("red", "blue")
plot(MSRP ~ Weight, data = car_data, main="MSRP by Weight + Hwy.MPG ",
    col=colors[factor(car_data$Sport)], pch=shapes[factor(car_data$Sport)])

abline(model_w , col="blue" , lwd=2)
abline(model_h , col="red" , lwd=2)

legend("topleft",
       legend = c("MSRP", "Weight + HWY.MPG"),
       pch = shapes,
       col = colors)
summary(model_w)
summary(model_h)
```
